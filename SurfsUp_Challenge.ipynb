```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

from scipy import stats
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June.
june_result = session.query(Measurement.tobs).filter(func.extract('month', Measurement.date) == 6).all()
```

```python
#  3. Convert the June temperatures to a list.
june_temperatures = [result for result, in june_result]
june_temperatures
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
# district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
june_temperatures_df = pd.DataFrame(june_temperatures, columns=['June Temps'])
june_temperatures_df.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temperatures_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_result = session.query(Measurement.tobs).filter(func.extract('month', Measurement.date) == 12).all()
```

```python
# 7. Convert the December temperatures to a list.
december_temperatures = [result for result, in december_result]
december_temperatures
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_temperatures_df = pd.DataFrame(december_temperatures, columns=['December Temps'])
december_temperatures_df.head()
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_temperatures_df.describe()
```

# D3: High-level summary of the temperature results

```python
# Using a T-Test, we can detrmine if the mean temperature values are different for the two months. To do so, we have to define our Null and Alternate Hypothesis.
# Null Hypothesis: june = december (the means of both populations are equal)
# Alternate Hypothesis: june  december (the means of both populations are not equal)
```

```python
# Important to note, we are specifying that the population does not have equal variance passing along False for the equal_var parameter. 
# We know this because both samples were taken from populations (june and december) with different standard deviations. 
#run independent sample T-Test
tStat, pValue = stats.ttest_ind(june_temperatures_df, december_temperatures_df, equal_var = False) 
#print the P-Value and the T-Statistic
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) 
```

```python
# The stats ttest_ind function runs the independent sample T-Test and outputs a P-Value and the Test-Statistic. 
# we have a very small P-value to reject the Null Hypothesis.
```
